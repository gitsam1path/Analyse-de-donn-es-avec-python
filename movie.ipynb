#Nous constatons qu il y a beaucoup de valeurs aberrantes pour gross et budget

```python
#Determinons le nombre exacte de valeurs aberrantes
df= data.select_dtypes(include="number")

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

Valeur_limite_inf = Q1 - 1.5 * IQR
Valeur_limite_sup = Q3 + 1.5 * IQR

((df <Valeur_limite_inf)| (df > Valeur_limite_sup)).sum()
```
